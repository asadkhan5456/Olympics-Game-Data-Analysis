```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
athletes=pd.read_csv('athlete_events.csv')
region=pd.read_csv('noc_regions.csv')
```

```python
athletes.head()
```

```python
region.head()
```

```python
df=athletes.merge(region, how='left',on= 'NOC')
df.head()
```

```python
df.rename(columns={'region':'Region','notes':'Notes'}, inplace= True);
```

```python
top_10_countries=df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
plt.figure(figsize=(12,6))
plt.title('Top 10 countries participating in Olympics')
sns.barplot(x=top_10_countries.index, y=top_10_countries, palette='Set2')
```

```python
plt.figure(figsize=(12,6))
plt.title('Age distribution of Athletes')
plt.xlabel('Age')
plt.ylabel('Number of Participants')
plt.hist(df.Age, bins=np.arange(10,80,2), color='blue', edgecolor='white')
```

```python
gender=df.Sex.value_counts()
gender
```

```python
plt.figure(figsize=(12,6))
plt.title('Gender Distribution')
plt.pie(gender, labels=gender.index, autopct='%1.1f%%', startangle=150, shadow=True)
```

```python
df.Medal.value_counts()
```

```python
# Total Number of Female Athletes

female_athletes= df[(df.Sex=='F') & (df.Season=='Summer')][['Sex','Year']]
female_athletes=female_athletes.groupby('Year').count().reset_index()
female_athletes.tail()
```

```python
women_athletes=df[(df.Sex=='F') & (df.Season=='Summer')]
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(20,10))
sns.countplot(x='Year',data=women_athletes, palette='Spectral')
plt.title('Women Representation')
```

```python
line=women_athletes.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
line.loc[:,'F'].plot()
plt.title('Line Plot of Female Athletes over the year')
```

```python
# Gold Medal

gold_medal=df[(df.Medal=='Gold')]
gold_medal.head()
```

```python
gold_medal=gold_medal[np.isfinite(gold_medal['Age'])]
```

```python
#Gold Medal above age of 60

gold_medal['ID'][gold_medal['Age']>60].count()
```

```python
sporting_event=gold_medal['Sport'][gold_medal['Age']>60]
sporting_event
```

```python
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title('Gold medal athletes above 60')
```

```python
gold_medal.Region.value_counts().reset_index(name='Medal').head(5)
```

```python
totalgold_medal=gold_medal.Region.value_counts().reset_index(name='Medal').head(5)
tot= sns.catplot(x='index',y='Medal',data=totalgold_medal, height=5, kind='bar',palette='rocket')
tot.despine(left=True)
tot.set_xlabels('Top 5 Countries')
tot.set_ylabels('Number of medals')
plt.title('Gold medal per country')
```

```python
#stats on Rio olympics

max_year=df.Year.max()
print(max_year)

teams=df[(df.Year==max_year) & (df.Medal == 'Gold')].Team

teams.value_counts().head(10)
```

```python
sns.barplot(x=teams.value_counts().head(20), y=teams.value_counts().head(20).index)
plt.ylabel(None);
plt.xlabel('Country with their medal in 2016')

```
